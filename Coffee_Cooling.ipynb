```python
from pylab import *
```

```python
def coffee_cooling(T_room, T_coffee, C, cream_time):
    dt = 1
    T_list = [0]*100
    T_list[0]=T_coffee
    for t in range(1,100):
        if t == cream_time:
            T_list[t-1]=T_list[t-1]-5
        T_list[t]=T_list[t-1]-((C*(T_list[t-1]-T_room))*dt)
    return T_list
```

```python
obs_data = [82.3,78.5,   74.3,   70.7,   67.6,  65. ,  62.5,  60.1, 58.1,56.1,   54.3, 52.8,  51.2,  49.9 ,48.6, 47.2, 46.1,  45. ,  43.9, 43.0, 41.9, 41.]
```

```python
temp_list = coffee_cooling(T_room=23, T_coffee=95, C=0.02, cream_time=1)
print temp_list
```

```python
def plot_coffee(temp_list, cream_insert_time):
    plot(temp_list[:15], label="coffee temp (Celsius)")
    plot(obs_data[:15], label="observed valued")
    plot([75]*15, label="optimal temperature")
    ylabel(r'Coffee temp (Celsius)')
    xlabel(r'Time (minutes)')
    annotate(r'Optimal temp', xy=(75, 75), xycoords='data', xytext=(+10, +50), textcoords='offset points', arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2'))
    legend(loc='upper right')
    title('Coffee cooling with cream insertion at time ' + str(cream_insert_time-1))
    show()
```

```python
def plot_mult_cream_insertions(T_room, T_coffee, C):
    for insert_time in range(1,15):
        cream_temp_list = coffee_cooling(T_room=T_room, T_coffee=T_coffee, C=C, cream_time=insert_time)
        plot_coffee(temp_list=cream_temp_list, cream_insert_time=insert_time)
        
```

```python
plot_mult_cream_insertions(23, 90, 0.03)
```
