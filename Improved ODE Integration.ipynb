#     Improved ODE Integration - Falling Ball And Harmonic Oscillator

### Setup

```python
from pylab import *
DELTA_T=0.1
def y_prime_euler(yn, vn, dt):
    return yn+vn*dt
def v_prime_euler(vn, dt, an=-9.8, yn=None):
    return vn+an*dt
```

```python
def solve_falling_ODE(x, dt, y_prime, v_prime, an=-9.8):
    # takes state x(y, velocity) -> returns state_prime(y_prime, v_prime)
    return array([y_prime(yn=x[0], vn=x[1], dt=dt), v_prime(vn=x[1], dt=dt)])
```

```python
def solve_spring_ODE(x, dt, y_prime, v_prime, k=1.0, m=1.0):
    return array([y_prime(yn=x[0], vn=x[1], dt=dt), v_prime(vn=x[1], an=-k*x[0]/m, dt=dt, yn=x[0])])
```

```python
def analytic_sol_falling(y0, t, g=9.8):
    return -0.5*g*(t**2)+y0
```

```python
def analytic_sol_spring(t, x_0=1.0,k=1.0, m=1.0):
    return x_0*math.cos(math.sqrt((k/m))*t)
```

```python
def gen_timesteps(start_range, end_range, dt):
    return arange(start_range, end_range, dt)
```

```python
analytic_falling_data=[]
for time_period in gen_timesteps(0, 10, DELTA_T):
    analytic_falling_data.append(analytic_sol_falling(y0=10, t=time_period))
    if analytic_falling_data[-1]<0:
        break
```

```python
plot(analytic_falling_data, label='analytic solution')
legend(loc='upper right')
xlabel('Number of timesteps, dt=%d s' % DELTA_T)
```

## Euler Richardson

### Falling Body

```python
def y_prime_euler_richardson(yn, vn, dt, an=-9.8):
    k=1.0
    m=1.0
    
    an=(-k*yn/m)
    v_mid=vn+0.5*an*dt
    return yn+v_mid*dt 
```

```python
def v_prime_euler_richardson(vn, dt, an=-9.8):
    a_mid=an #acceleration is constant
    return vn+a_mid*dt
```

```python
state = [array([10,0])]
while True:
    state.append(solve_falling_ODE(x=state[-1], dt=DELTA_T, y_prime=y_prime_euler_richardson, v_prime=v_prime_euler_richardson))
    if state[-1][0]<0:
        break
state = array(state)
state = state[:,0]
print state
```

### Harmonic Oscillation

##### Euler-Richardson

```python
def v_prime_euler_richardson_spring(vn, dt, an, yn=None, k=1.0, m=1.0):
    v_mid = vn+0.5*an*dt
    y_mid = yn+0.5*vn*dt
    a_mid = -k*y_mid/m
    return vn+a_mid*dt
```

```python
state = [array([1.0,0.0])]
state_euler = [array([1.0,0.0])]
period_count=0
while True:
    state.append(solve_spring_ODE(x=state[-1], dt=DELTA_T, y_prime=y_prime_euler_richardson, v_prime=v_prime_euler_richardson_spring))
    state_euler.append(solve_spring_ODE(x=state_euler[-1], dt=DELTA_T, y_prime=y_prime_euler, v_prime=v_prime_euler))
    if (state[-1][0]<0 and state[-2][0]>0) or (state[-1][0]>0 and state[-2][0]<0):
        period_count+=1
    if period_count > 10:
        break
state = array(state)
state = state[:,0]
state_euler = array(state_euler)
state_euler = state_euler[:,0]
#print state

```

```python
plot(state, label="Euler-Richardson simulation")
plot(state_euler, label="Euler method simulation")
xlabel("Number of timesteps, dt=%f s" % DELTA_T)
ylabel("x position")
legend(loc='lower left')
```

## Runge Kutta Methods

```python
def y_prime_runge_kutta(yn, vn, dt):
    k=1.0
    m=1.0
    an = (-k*yn/m)
    k1 = yn+an*dt
    k2 = (yn+k1/2)*an*dt
    k3 = (yn+k2/2)*an*dt
    k4 = (yn+k3)*an*dt
    return yn+(k1+2*k2+2*k3+k4)/6
```

```python
def v_prime_runge_kutta(vn, dt, an, yn=None, k=1.0, m=1.0):
    k1 = -k*yn/m*dt
    k2 = -k*(yn+k1/2)/m*dt
    k3 = -k*(yn+k2/2)/m*dt
    k4 = -k*(yn+k3)/m*dt
    return yn+(k1+2*k2+2*k3+k4)/6
```

```python
state = [array([1.0,0.0])]
state_euler = [array([1.0,0.0])]
state_rk = [array([1.0, 0.0])]
period_count=0
while True:
    state_rk.append(solve_spring_ODE(x=state[-1], dt=DELTA_T, y_prime=y_prime_runge_kutta, v_prime=v_prime_runge_kutta))
    state.append(solve_spring_ODE(x=state[-1], dt=DELTA_T, y_prime=y_prime_euler_richardson, v_prime=v_prime_euler_richardson_spring))
    state_euler.append(solve_spring_ODE(x=state_euler[-1], dt=DELTA_T, y_prime=y_prime_euler, v_prime=v_prime_euler))
    if (state[-1][0]<0 and state[-2][0]>0) or (state[-1][0]>0 and state[-2][0]<0):
        period_count+=1
    if period_count > 10:
        break
state = array(state)
state = state[:,0]
state_euler = array(state_euler)
state_euler = state_euler[:,0]
state_rk = array(state_rk)
state_rk = state_rk[:,0]
#print state

plot(state, label="Euler-Richardson simulation")
plot(state_euler, label="Euler method simulation")
plot(state_rk, label="Runge-Kutta")
xlabel("Number of timesteps, dt=%f s" % DELTA_T)
ylabel("x position")
legend(loc='lower left')
```
