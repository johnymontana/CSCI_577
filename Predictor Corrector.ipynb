```python
# import packages
from scipy import *
from numpy import *
from matplotlib.pyplot import *
# from __future__ import division
 
###### Global variables:
m = 1.			# mass attached to spring
k = 1.  		# spring constant 
x0 = 1. 		# initial position is 1 meter
v0 = 0. 		# initial velocity is zero meters/second
 
# time step and number of time steps required for 5 periods
tSt = 0.2
cSt = int(math.ceil(10*math.pi/(math.sqrt(k/m)*tSt)))
 
# Predictor
# input:	
#	rgx:	an array containing the position and velocity (at step n-1)
#	rgv:	an array containing the velocity and acceleratoin (at step n)
#	tStep:	time step
# output:
#	an array containing the predicted position and velocity
def rgPredict(rgx, rgv, tStep):
	return rgx + 2*rgv*tStep
 
# Corrector
# input:
#	rgx:	an array containing the position and velocity (at step n)
#	rgv:	an array containing the velocity and acceleration (at step n)
#	rgP:	an array containing the predicted velocity and acceleration
#	tStep:	time step
def rgCorrect(rgx, rgv, rgP, tStep):
	return rgx + (rgv + rgP)*tStep/2
 
# Runge Kutta Method
# input: 
#	f:	function
#	t:	time variable (at step n)
#	x:	position variable (at step n)
#	dt:	time step
# output:
#	the next position based on the current position
def rgRungeKutta(f, t, x, dt):
    k1 = f(x, t)*dt
    k2 = f(x + (k1/2),t + (dt/2))*dt
    k3 = f(x + (k2/2), t + (dt/2))*dt
    k4 = f(x + k3, t + dt)*dt
    return x + (1/6.)*(k1 + 2*k2 + 2*k3 + k4)
 
# Spring approximate solution
# input:
#	y:	a vector including the position and velocity at some point in time
# output:
#	an array containting the derivative evaluated at that point
def rgSpring(y, t):
	return array([y[1], -k/m*y[0]]) 
 
rgt = [0, tSt]
rgStateC = [array([x0, v0])] 
rgStateC.append(rgRungeKutta(rgSpring, rgt[0], rgStateC[-1], tSt))
 
for iSt in range(0, cSt):
	rgt.append((iSt+1)*tSt)
	rgStateP = rgPredict(rgStateC[-2], rgSpring(rgStateC[-1], rgt[-1]), tSt)
	rgStateC.append(rgCorrect(rgStateC[-1], rgSpring(rgStateC[-1], rgt[-1]), rgSpring(rgStateP, rgt[-1]), tSt))
 
rgPos = array(rgStateC)[:, 0]
rgVel = array(rgStateC)[:, 1]
 
plot(rgt, rgPos, label = "Position")
plot(rgt, rgVel, label = "Velocity")
 
xlabel('Time (note: dt = %.2f)' %tSt)
ylabel('Amplitude from Equilibrium (0.0)')
title('Simple Harmonic Oscillator')
grid(True)
legend()
show()
```
