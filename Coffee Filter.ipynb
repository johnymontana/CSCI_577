```python
# import packages
from __future__ import division
from scipy import * 
from numpy import * 
from matplotlib.pyplot import * 
 
# Global Variables
dAccGrav = 9.81
dVelTerm = -0.5
 
# Finite Difference
# input:
#	rgdPos:	array of positions
#	rgt:	array of times
# output:
#	the velocity and acceleration at the position
def fnrgFinDiff(rgdPos, rgt):
	rgdVel = (rgdPos[2:] - rgdPos[:-2])/(rgt[2:] - rgt[:-2])
	rgdAcc = (rgdPos[2:] - 2*rgdPos[1:-1] + rgdPos[:-2])/(0.5*(rgt[2:] - rgt[:-2]))**2
	return (rgdVel, rgdAcc)
 
# Time in seconds: rgData[0,:]
# y-axis position in meters: rgData[1,:]
rgData = array([[.2055,.2302,.2550,.2797,.3045,.3292,.3539,.3786,.4033,.4280,
				.4526,.4773,.5020,.5266,.5513,.5759,.6005,.6252,.6498,.6744,
				.6990,.7236,.7482,.7728,.7974,.8220,.8466],
				[.4188,.4164,.4128,.4082,.4026,.3958,.3878,.3802,.3708,.3609,
				.3505,.3400,.3297,.3181,.3051,.2913,.2788,.2667,.2497,.2337,
				.2175,.2008,.1846,.1696,.1566,.1393,.1263]])
 
rgt = rgData[0, :]
rgdPos = rgData[1, :]
rgdVel = []
rgdAcc = []
 
(rgdVel, rgdAcc) = fnrgFinDiff(rgdPos, rgt)
 
subplot(221)
plot(rgt, rgdPos, 'ro-')
xlabel('Time (s)')
ylabel('Position (m)')
title('Position Vs. Time')
subplot(222)
plot(rgt[1:-1], rgdVel, 'go-')
xlabel('Time (s)')
ylabel('Velocity (m/s)')
title('Velocity Vs. Time')
subplot(223)
plot(rgt[1:-1], rgdAcc,'ko-')
xlabel('Time (s)')
ylabel('Acceleration (m/s^2)')
title('Acceleration Vs. Time')
subplot(224)
plot(rgdVel, rgdAcc,'o') 
xlabel('Velocity (m/s)')
ylabel('Acceleration (m/s^2)')
title('Velocity Vs. Acceleration')
show()
```

```python
# Runge Kutta Method
# input: 
#	fn:		function
#	t:		time variable (at step n)
#	dPos:	position variable (at step n)
#	tStep:	time step
# output:
#	the next position based on the current position
def fnrgRungeKutta(fn, dPos, tStep):
	k1 = fn(dPos)*tStep
	k2 = fn(dPos + (k1/2))*tStep
	k3 = fn(dPos + (k2/2))*tStep
	k4 = fn(dPos + k3)*tStep
	return dPos + (1/6)*(k1 + 2*k2 + 2*k3 + k4)
 
# Falling body approximate solution (linear)
# input:
#	rgState:	a vector including the position and velocity of the body at some point in time
# output:
#	an array containting the derivative evaluated at that point
def fnrgFallLin(rgState):
	return array([rgState[1], -dAccGrav*(1 - rgState[1]/dVelTerm)])
 
# Falling body approximate solution (quadratic)
# input:
#	rgState:	a vector including the position and velocity of the body at some point in time
# output:
#	an array containting the derivative evaluated at that point
def fnrgFallQuad(rgState):
	return array([rgState[1], -dAccGrav*(1 - (rgState[1]/dVelTerm)**2)])
 
# Approximate solutions 
rgdStateLin = [array([rgdPos[0], 0])]		# assuming no intial velocity, may change to assume 
rgdStateQuad = [array([rgdPos[0], 0])]		# initial velocity found via finite difference
 
for it in rgt[0:-1]:
	rgdStateLin.append(fnrgRungeKutta(fnrgFallLin, rgdStateLin[-1], rgt[it+1] - rgt[it]))
	rgdStateQuad.append(fnrgRungeKutta(fnrgFallQuad, rgdStateQuad[-1], rgt[it+1] - rgt[it]))
 
rgdStateLin = array(rgdStateLin)
rgdStateQuad = array(rgdStateQuad)
 
figure(2)
plot(rgt, rgdStateLin[:, 0], 'ro-', label = 'Linear Simulation')
plot(rgt, rgdStateQuad[:, 0], 'go-', label = 'Quadratic Simulation')
plot(rgt, rgdPos, 'ko-', label = 'Empirical Data')
title('Simulated and Empirical Positions')
xlabel('Time (s)')
ylabel('Position (m)')
legend() 
show()
```
